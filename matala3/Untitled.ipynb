```python
file=open("�צאט WhatsApp עם יום הולדת בנות לנויה.txt",encoding='utf8')       
def chat(file):
    import json
    lst = []
    user= dict()
    id_num = 0
    final_json= dict()
    for line in file:
            exists = None
            try:
                if ("<המדיה לא נכללה>" in line or "הוסיף/ה" in line
                        or "ההודעות והשיחות מוצפנות מקצה-לקצה" in line or "נוצרה על ידי" in line):
                    continue
                txt = dict()
                date_index = line.index('-')
                date = line[0:date_index]
                filter_name = line[date_index + 1:line.index(':', date_index)]
                text = line[line.index(':', date_index) + 1:]
                for key, value in user.items():
                    if value == filter_name:
                        exists= key
                if exists is not None:
                    filter_name= exists
                else:
                    user[id_num] =filter_name
                    filter_name=id_num
                    id_num += 1
                txt['datetime']=date
                txt['id']=filter_name
                txt['text'] = text
                lst.append(txt)
            except:
                lst[-1]['text'] += line


    file=open("�צאט WhatsApp עם יום הולדת בנות לנויה.txt",encoding='utf8')       
    for line in file:
        if not  "נוצרה על ידי"  in line:
            continue
        metadata = dict()
        chatname_index = line.split('"')
        chatname = chatname_index[1]
        metadata["chat_name"] = chatname
        creationdate = line.split('-')
        metadata["creation_date"]=creationdate[0]
        metadata["num_of_participants"]=id_num+1
        creator_index=line.index('ידי')
        metadata["creator"]=line[creator_index+6:creator_index+22]

    final_json["messages"] = lst 
    final_json["metadata"] = metadata
    file_in_json = chatname+".txt"
    outputfile = open(file_in_json, 'w', encoding='utf8')
    outputfile.write(json.dumps(final_json, indent=4, ensure_ascii=False))
    outputfile.close()
    return(final_json)
chat(file)
```
